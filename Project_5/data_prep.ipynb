```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_russell = pd.read_csv("./data/stockMarketCloseValuesRussell1000.csv")
data_russell = data_russell.iloc[: , 1:] # Remove the index column

data_russell.dropna(axis=1, inplace=True) # Remove columns with NaN 

buy = data_russell.iloc[0 : -1 ].reset_index(drop=True) # Buy at time t
sell = data_russell.iloc[1 : ].reset_index(drop=True) # Sell at time t+1

returns_russell = (sell)/buy # Returns as calculated in problem statement


returns_russell = returns_russell.round(6)

returns_russell.to_csv("./data/returns_russell.csv")
returns_russell
```

```python
data_wilshire = pd.read_csv("./data/stockMarketCloseValueswilshire5000.csv")
data_wilshire = data_wilshire.iloc[: , 3:] # Drop index, w5000 and Unnamed

data_wilshire.dropna(axis=1, inplace=True) # Drop columns with NaN

buy = data_wilshire.iloc[0 : -1 ].reset_index(drop=True)
sell = data_wilshire.iloc[1 : ].reset_index(drop=True)
returns_wilshire = (sell)/buy

returns_wilshire = returns_wilshire.round(6)

returns_wilshire.to_csv("./data/returns_wilshire.csv")
returns_wilshire
```

```python
russell_cols_to_take = returns_russell.columns.difference(returns_wilshire.columns)

merged_returns = pd.merge(returns_russell[russell_cols_to_take], returns_wilshire, left_index=True, right_index=True, how='outer')

merged_returns.rename( columns = lambda x : x.replace('.Close', ''), inplace=True )

merged_returns
```

```python
merged_returns.to_csv("./data/combined_returns.csv", index=False)
```

```python
import time

merged_returns = pd.read_csv("./data/combined_returns.csv")
combined_returns = merged_returns.iloc[ : 1000  , : 2000]
combined_returns_cols_arr = np.array(combined_returns.columns)
combined_returns_arr = np.array(combined_returns, dtype=np.float16)

num_assets = len(combined_returns_cols_arr)

t1 = time.time()
print("Starting to compute pairs and pairs names")

i, j = np.triu_indices(num_assets, k=1)
all_pairs_names = np.column_stack((combined_returns_cols_arr[i], combined_returns_cols_arr[j]))
all_pairs_indices = np.column_stack((i, j))
num_pairs = len(all_pairs_names)

t2 = time.time()
print(f"Done with computing all pairs in time: {t2 - t1} seconds")


pair_indices = all_pairs_indices.T
t3 = time.time()
print("Starting to compute all deltas")
deltas = (combined_returns_arr[:,all_pairs_indices[:, 0]] - combined_returns_arr[:,all_pairs_indices[:, 1]])
deltas = deltas.astype(np.float16)

t4 = time.time()
print(f"Done with all deltas in : {t4-t3} seconds")
```

```python
def evalGrad(x, rbar, theta, T, pi, rt):

    # First term of gradient
    q1 = -rbar

    # Second term of gradient
    q2 = (theta * pi / T) * np.sum( (( np.multiply((rt - rbar) , x) )).T @ (rt-rbar), axis=0)

    return q1 + q2

x = np.array( [ 1/num_pairs for _ in range(num_pairs) ] , dtype = np.float32)

evalGrad( x, np.mean(deltas, axis=0), 5, 1000, 2, deltas)
```
